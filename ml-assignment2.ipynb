```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 20)
y_true = 2 * x + 1
y_pred = y_true + np.random.normal(0, 1.5, 20)

absolute_errors = np.abs(y_true - y_pred)
squared_errors = (y_true - y_pred) ** 2

mae = np.mean(absolute_errors)
mse = np.mean(squared_errors)

print("Individual Absolute Errors:", absolute_errors)
print("Mean Absolute Error (MAE):", mae)
print("Individual Squared Errors:", squared_errors)
print("Mean Squared Error (MSE):", mse)

plt.figure(figsize=(10, 6))
plt.scatter(x, y_true, label='True Values')
plt.scatter(x, y_pred, label='Predicted Values')
plt.plot(x, y_true, 'b-', label='True Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('True vs Predicted Values')
plt.legend()
plt.grid(True)
plt.show('predictor_error_plot.png')
plt.close()
```
